```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y(x,t):
    y=24*(np.cos(((2*np.pi)/48)*(x-6*t)))
    return y
    
```

```python
fig = plt.figure()
x = np.linspace(0,96,50)

# plot the function
plt.plot(x,y(x,0),label='t=0')
plt.plot(x,y(x,1), label='t=1')
plt.plot(x,y(x,2), label='t=2')
plt.plot(x,y(x,3), label='t=3')
plt.plot(x,y(x,4), label='t=4')

plt.legend(loc='upper left')

# show the plot
plt.show()
```
